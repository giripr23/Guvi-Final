## Data pre processing
## Remove the row, if any column in the 
## list of columns[ "num_subscribers", "num_reviews", "num_lectures", "content_duration ]
## contains 0 or less value

```python
for col in ["num_subscribers", "num_reviews", "num_lectures", "content_duration"]:
    a = df[col] <= 0
    df.loc[a] = np.nan
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.head(20)
```

```python
df["year"] = pd.to_datetime(df["published_timestamp"]).dt.year
df["month"] = pd.to_datetime(df["published_timestamp"]).dt.month
```

```python
df = df.drop(["course_title", "url", "published_timestamp"], axis = 1)
df.head(10)
```

```python
X=df.iloc[:,1:5] ## all rows, from column1 to Column 4.
## That is from column price to column num_lectures
y=df['Rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

## multiple linear regression
regr = LinearRegression()
fit = regr.fit(X_train, y_train)

print(regr.coef_)
print(regr.intercept_)
```

```python
print(regr.score(X_test, y_test))
```

```python
y_pred = regr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)

print("R2 score:", r2)
```
